```python
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",names=['sepal length','sepal width','petal length','petal width','target'])
df.head(10)
```

```python
v=sns.pairplot(df,kind='scatter',hue='target')
v
plt.show()
```

```python
correlation=df.corr()
plt.figure(figsize=(7,6))
sns.heatmap(correlation,annot=True)
plt.show()
```

```python
X=df.iloc[:,:4]
X.head(10)
```

```python
y=df.iloc[:,4:]
y.head(10)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=20,random_state=0)
```

```python
X.shape,y.shape,X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_test,X_train
```

```python
clf = DecisionTreeClassifier(random_state=10)
model = clf.fit(X_train, y_train)
```

```python
predict=model.predict(X_test)
```

```python
print('accuracy score of the model:')
print(accuracy_score(y_test,predict)*100)
print("clasifiction report:")
print(classification_report(y_test,predict))
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(20,10))
plot_tree(clf,filled=True)
plt.show()
```
